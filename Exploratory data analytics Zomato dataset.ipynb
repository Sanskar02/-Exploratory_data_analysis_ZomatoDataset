```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts()
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
plt.pie(country_val[:3], labels = country_names[:3], autopct='%1.2f%%')
```

Observation : Zomato maximum records or transaction are from India
After that USA and then United Kingdoms 

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

## Observations
1. When rating is between 4.5 to 4.9 ----> Excellent
2. When rating is between 4.0 to 3.4 ----> very good
3. When rating is between 3.0 to 3.4 ----> good
4. When rating is between 2.5 to 2.9 ----> average
5. When rating is between 2.5 to 2.9 ----> average
6. When rainhg is between 2.0 to 2.4 ----> poor

```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating', y='Rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green'])
```

Observations:

1. Not rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
### Count Plot
import seaborn as sns
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','black'])
```
